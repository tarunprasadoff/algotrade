```python
#!/usr/bin/env python 

import csv
import datetime
import re

import requests
import pandas as pd
import arrow
import datetime

import mplfinance as mpf
import matplotlib.pyplot as plt
```

```python
def get_quote_data(symbol='SBIN.NS', data_range='1d', data_interval='1m'):
    res = requests.get('https://query1.finance.yahoo.com/v8/finance/chart/{symbol}?range={data_range}&interval={data_interval}'.format(**locals()))
    data = res.json()
    body = data['chart']['result'][0]    
    dt = datetime.datetime
    dt = pd.Series(map(lambda x: arrow.get(x).to('Asia/Calcutta').datetime.replace(tzinfo=None), body['timestamp']), name='Datetime')
    df = pd.DataFrame(body['indicators']['quote'][0], index=dt)
    dg = pd.DataFrame(body['timestamp'])    
    df = df.loc[:, ('open', 'high', 'low', 'close', 'volume')]
    df.dropna(inplace=True)     #removing NaN rows
    df.columns = ['Open', 'High','Low','Close','Volume']    #Renaming columns in pandas
    df.index.name = 'Date'
    
    return df

data = get_quote_data('DABUR.NS', '2d', '15m')
print(data)
```

```python
def style():
    mc = mpf.make_marketcolors(
                            up='tab:green',down='tab:red',
                            edge='black',
                            wick={'up':'green','down':'red'},
                            volume={'up':'green','down':'red'},
                           )

    return mpf.make_mpf_style(marketcolors=mc)

def bollvwap(data, spanb, spanv, stds):
    num = (((data['High']+data['Low']+data['Close'])/3)*data['Volume']).rolling(spanv).sum()
    den = data['Volume'].rolling(spanv).sum()
    return [mpf.make_addplot(data.Close.rolling(spanb).mean(),color='black'),mpf.make_addplot(data.Close.rolling(spanb).mean()+stds*data.Close.rolling(spanb).std(),color='black'),mpf.make_addplot(data.Close.rolling(spanb).mean()-stds*data.Close.rolling(spanb).std(),color='black'),mpf.make_addplot(num/den, color='blue')]

def plotatr(data, span):
    a = pd.DataFrame([data['High']-data['Low'],abs(data['High']-data['Close'].rolling(2).sum()+data['Close']),abs(data['Low']-data['Close'].rolling(2).sum()+data['Close'])]).max(axis = 0).rolling(span).mean()
    fig = plt.figure()
    fig.set_figheight(20)
    fig.set_figwidth(20)
    plt.plot(a)
    #plt.figure(figsize=(50,50))
```

```python
mpf.plot(data,addplot=bollvwap(data, 10, 5, 2),type='candle',volume=True,style=style(),figratio=(20,20))
plotatr(data,14)
```
