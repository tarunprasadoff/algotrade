```python
def style():
    mc = mpf.make_marketcolors(
        up='tab:green', down='tab:red',
        edge='black',
        wick={'up': 'green', 'down': 'red'},
        volume={'up': 'green', 'down': 'red'},
    )

    return mpf.make_mpf_style(marketcolors=mc)


def boll(ndata):
    return [mpf.make_addplot(ndata.MBB, color='black'), mpf.make_addplot(ndata.UBB, color='black'), mpf.make_addplot(ndata.LBB, color='black')]


def dayplot(rdata):
    ndata = rdata.copy()
    ndata.rename(columns={'Datetime': 'Date'}, inplace=True)
    ndata.set_index('Date', inplace=True)
    ndata.index = pd.to_datetime(ndata.index)
    mpf.plot(ndata, addplot=boll(ndata), type='candle',
             volume=True, style=style(), figratio=(20, 20))
```

```python
def daydata(ndata, ndate):
    i = 0
    while(True):
        if(ndata['Datetime'][ndata.index == i].values[0][0:10] == ndate):
            break
        else:
            i += 1
    st = i

    if(ndate == ndata.tail(1).Datetime.values[0][0:10]):
        en = len(ndata)
    else:
        while(True):
            if(ndata['Datetime'][ndata.index == i].values[0][0:10] != ndate):
                break
            else:
                i += 1
                en = i

    return ndata[st:en]
```

```python
def simulate(data):

    temp = data
    templag1 = data.shift(1)
    templag2 = data.shift(2)
    templead1 = data.shift(-1)

    templag1mean = templag1[['Open', 'Close']].mean(axis=1)

    condition1 = (temp.Close > temp.Open) & (templag2.Close < templag2.Open)
    condition2 = (temp.Open > templag1mean) & (
        temp.Close > templag2.Open) & (templag2.Close > templag1mean)
    condition3 = (temp.Close < (1 * temp.MBB + 0 * temp.LBB))

    length = len(temp.Datetime[0])
    daytime = temp.Datetime.str[length-8:length]
    condition4 = daytime > '10:30:00'
    condition5 = daytime < '12:30:00'

    data['signals'] = condition1 & condition2 & condition3 & condition4 & condition5

    unitsuccess = temp.Close * .005

    out1 = (data.shift(-2).Close - templead1.Open) > unitsuccess
    out2 = (data.shift(-3).Close - templead1.Open) > unitsuccess
    out3 = (data.shift(-4).Close - templead1.Open) > unitsuccess
    out4 = (data.shift(-5).Close - templead1.Open) > unitsuccess
    out5 = (data.shift(-6).Close - templead1.Open) > unitsuccess
    out6 = (data.shift(-7).Close - templead1.Open) > unitsuccess
    out7 = (data.shift(-8).Close - templead1.Open) > unitsuccess
    out8 = (data.shift(-9).Close - templead1.Open) > unitsuccess
    out9 = (data.shift(-10).Close - templead1.Open) > unitsuccess
    out10 = (data.shift(-11).Close - templead1.Open) > unitsuccess
    out11 = (data.shift(-12).Close - templead1.Open) > unitsuccess
    out12 = (data.shift(-13).Close - templead1.Open) > unitsuccess
    out13 = (data.shift(-14).Close - templead1.Open) > unitsuccess
    out14 = (data.shift(-15).Close - templead1.Open) > unitsuccess
    out15 = (data.shift(-16).Close - templead1.Open) > unitsuccess
    out16 = (data.shift(-17).Close - templead1.Open) > unitsuccess
    out17 = (data.shift(-18).Close - templead1.Open) > unitsuccess
    out18 = (data.shift(-19).Close - templead1.Open) > unitsuccess
    out19 = (data.shift(-20).Close - templead1.Open) > unitsuccess
    out20 = (data.shift(-21).Close - templead1.Open) > unitsuccess
    out21 = (data.shift(-22).Close - templead1.Open) > unitsuccess
    out22 = (data.shift(-23).Close - templead1.Open) > unitsuccess
    out23 = (data.shift(-24).Close - templead1.Open) > unitsuccess

    data['good'] = (out1 | out2 | out3 | out4 | out5 | out6 |
                    out7 | out8 | out9 | out10 | out11 | out12 | out13 | out14 | out15 | out16 | out17 |
                    out18 | out19 | out20 | out21 | out22 | out23) & data['signals']

    good = data.loc[data.good == True].shape[0]
    bad = data.loc[data.signals == True].shape[0] - good

    return good, bad
```

```python
netgood = 0
netbad = 0
for name in company:
    print(name)
    good = 0
    bad = 0
    data = pd.read_csv(
        'C:/Users/user/Stockomation/Github/Data/5MinLatest/STOCK_DATA/'+name+'.csv')
    good, bad = simulate(data)

    print('Good Signals:' + str(good))
    print('Bad Signals:' + str(bad))
    netgood += good
    netbad += bad

print('The net result is:')
print('Net Good Signals:' + str(netgood))
print('Net Bad Signals:' + str(netbad))
```
